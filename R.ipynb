```r
x<-c(1,3,5,8,13,16)
y<-c(800000,850000,1000000,1300000,2100000,2600000)
```

```r
plot(x,y,col="blue", main="Years after 1990 versus Cost of Super Bowl Commercial",xlab="Years after 1990", ylab="Cost of Commercial")
```

```r
relation_nl<-nls(y~I(c*exp(k*x)), start=list(c=500000,k=0.1))
print(relation_nl)
summary(relation_nl)
```

```r
x_seq <- seq(from=1,to=20, by=0.01)
new_relation_nl<-list(0)
for (i in 1:length(x_seq)) {
    x_new <- data.frame(x=x_seq[i])
    new_relation_nl[i]<-predict(relation_nl, x_new)
}
plot(x,y,col="blue", main="Nonlinear Regression Example",lines(x_seq, new_relation_nl),xlab="x",ylab="y")
```
